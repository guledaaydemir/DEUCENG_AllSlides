```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
from sklearn.datasets import (load_boston)
boston_dataset = load_boston()
boston = pd.DataFrame(boston_dataset.data)
boston.columns = boston_dataset.feature_names
boston['PRICE'] = boston_dataset.target

from sklearn.linear_model import LinearRegression
X = boston.drop('PRICE', axis=1)
lm = LinearRegression()
lm.fit(X,boston.PRICE)

print('intercept', lm.intercept_)
print('coefficients', lm.coef_)
```

```python
mse = np.mean((boston.PRICE - lm.predict(X))**2)
print(mse)
```

```python
plt.scatter(boston.RM, boston.PRICE)
plt.xlabel('avg num of RM')
plt.ylabel('housing price')
plt.title('relationship btw RM and price')
plt.show()
```

```python
plt.scatter(boston.PRICE, lm.predict(X))
plt.xlabel('Prices')
plt.ylabel('predicted prices')
plt.title('relationship btw prices and predicted prices')
plt.show()
```
